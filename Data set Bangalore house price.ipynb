Importing the Libararies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
```

Read the data

```python
data = pd.read_csv("bangalore house price prediction OHE-data.csv")
```

Spliting the dataset into feature & Target

```python
x = data.drop(columns="price")
y = data["price"]
```

```python
x
```

```python
y
```

Spliting the dataset into train_test_split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x.shape,x_train.shape,x_test.shape
```

```python
model = KNeighborsRegressor(n_neighbors=3)
print("Model is loaded")
```

```python
model_fitted = model.fit(x_train,y_train)
print("Train the model")
```

```python
model_pred = model_fitted.predict(x_test)
```

```python
model_pred
```

Checking the Accuracy of the train & test data

```python
y_test
```

```python
acc_train = model_fitted.score(x_train,y_train)
acc_train
```

```python
acc_test = r2_score(y_test,model_pred)
acc_test
```
